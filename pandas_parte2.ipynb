# Tratamento dos Dados

## Remoção de Dados
Algunas dados podem não interessar na hora de realizar uma análise de dados. Além disso, podem existir __dados faltantes__ que são indicados por _"NaN"_ ou _"na"_. A ausência desses dados pode provocar erros no futuro ou então prejudicar modelos de aprendizagem de máquina.

```python
  .drop([columns], axis=, inplace=)
```
onde:

```python
  columns: são as colunas (_variables_)
  axis: são os eixos (1: coluna, 0 linha)
  inplace: True, altera o DataFrame definitivamente
```


````python
# inspecionar novamente o DataFrame
df.sample(5)
````

````python
# Eliminar dados que não afetam nosso DataFrame: Province/State
df.drop(['Province/State'], axis=1, inplace=True)
````

````python
# Verificando o resultado
df.head()
````

````python
df.info()
````

## Lidando com dados faltantes (_missing values_ `NA`)

| Method         | Description                                                  |
| -------------- | ------------------------------------------------------------ |
| `df.dropna()`  | Filtra `NA`, selecionar linhas (`axis = 0`) ou selecionar colunas (`axis=1`) |
| `df.fillna()`  | Preencher os `NA` com algum valor  |
| `df.isnull()`  | Retorna os valores booleanos indicando quais valores são `NA` |

````python
# verificar dados NA
df.dropna(axis=0).sample(10)
````

````python
# Existem nulos?
df.isnull()
````

````python
# Preencher com zero qualquer valor "na"
df.fillna(0, inplace=True)
````

## Salvando o trabalho feito até o momento

Quando carregamos os dados, eles foram trazidos de um local externo (Web). Agora vamos salvar os dados tratados em um arquivo local.

````python
# index = None indica que não salvaremos os índices criados pelo Pandas
df.to_csv('covid_tratado.csv', index=None)
````
